```python
import pandas as pd
from kmodes.kprototypes import KPrototypes
```

```python
df=pd.read_csv('F:/Course/youtube/4. K Means dan HClust/dataku.csv')
df
```

```python
df=df.drop(['CustomerID'],axis=1)
df
```

```python
my_array=df.values
my_array[:,1]=my_array[:,1].astype(float)
my_array[:,2]=my_array[:,2].astype(float)
my_array[:,3]=my_array[:,3].astype(float)
my_array
```

```python
model=KPrototypes(n_clusters=3,verbose=2,max_iter=5)
klaster=model.fit_predict(my_array,categorical=[0])
```

```python
print(model.cluster_centroids_)
```

```python
cluster_dict=[]
for c in klaster:
    cluster_dict.append(c)
```

```python
df['cluster']=cluster_dict
df
```
